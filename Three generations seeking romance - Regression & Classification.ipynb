## Feature Engineering

```python
df=df.drop(["essay0","essay1","essay2","essay3","essay4","essay5","essay6","essay7","essay8","essay9"],axis=1)
#We are deleting some columns.
```

```python
df["body_type"]=df["body_type"].map({"average":"fit","fit":"fit",
              "athletic":"fit",
              "thin":"fit",
              "curvy":"curvy",
              "a little extra":"curvy",
              "skinny":"fit",
              "full figured":"curvy",
              "overweight":"bad",
              "jacked":"bad",
              "used up":"bad",
              "rather not say":"curvy"
             })
```

```python
status={"mostly":"","strictly":""}
df["diet"].replace(status,regex=True,inplace=True)
df["diet"]=df["diet"].str.strip()
```

```python
df["education"].replace({"working on":"student","graduated from":"graduated","dropped out":"dropped"},regex=True)
df["education"]=df["education"].str.split().str[0]
df["education"].replace({"college/university":"student","two-year":"student","masters":"student","working":"student","space":"student","ph.d":"graduated","law":"graduated","high":"graduated"},regex=True,inplace=True)
```

```python
df.drop(["ethnicity","job"],axis=1,inplace=True)
```

```python
df["last_online_year"]=2021-(df['last_online'].str[:4]).astype(int)
df["last_online_month"]=(df['last_online'].str[5:7]).astype(int)
df.drop('last_online',axis=1)
```

```python
df.location=df.location.str.split(",").str[0]
```

```python
offsprings=[]
for i in df.offspring.fillna("null").str.replace("[^\w\s]",""):
    if "doesnrsquot" in i:
        offsprings.append("0")
    elif "null" in i:
        offsprings.append("0")
    else:
        offsprings.append("1")
        
df["offspring"]=offsprings
df["offspring"]=df["offspring"].astype(int)
```

```python
df["religion"]=df["religion"].str.split().str[0]
```

```python
df["sign"]=df["sign"].str.split().str[0]
```

```python
df["number_of_fluent_languages"]=df.speaks.str.count("fluently")
df["number_of_fluent_languages"].replace(0,1,regex=True,inplace=True)
df["number_of_poorly_languages"]=df.speaks.str.count("poorly")
df.drop("speaks",axis=1,inplace=True)
```

```python
for column in df.select_dtypes(include=np.number):
    df[column].fillna(df[column].median(),inplace=True)
#We fill all number values with average values.
```

```python
for column in df.select_dtypes(include=np.object):
    df[column].fillna(df[column].mode()[0],inplace=True)
#We fill all text values with the most repeated values.
```

```python
df.isnull().sum()
```

```python
df["generation"]=df["age"]
```

```python
df["generation"]
```

```python
for i in range(len(df["generation"])):
    if df["generation"][i]<33:
        df["generation"][i]=0
    elif df["generation"][i]<48:
        df["generation"][i]=1
    else:
        df["generation"][i]=2
df.generation.value_counts()
```

```python
abs(df.corr()['age']).sort_values(ascending=False)
```

## Regression

```python
x,y=df.drop(["age"],axis=1),df[["age"]]
x=pd.get_dummies(x,drop_first=True)
```

```python
x=scaler.fit_transform(x)
```

```python
pd.DataFrame(x).head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(800, activation='relu'))
model.add(Dense(800, activation='relu'))
model.add(Dense(800, activation='relu'))
model.add(Dense(800, activation='relu'))
model.add(Dense(800, activation='relu'))
model.add(Dense(800, activation='relu'))
model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=128,epochs=100,verbose=1)
model.summary()
```

```python
tahmin=model.predict(x_test)
```

```python
r2_score(tahmin,y_test)
```

```python
(mean_squared_error(tahmin,y_test))**0.5
```

## Classification

```python
a,b=df.drop(["generation"],axis=1),df[["generation"]]
a=pd.get_dummies(a,drop_first=True)
```

```python
a
```

```python
a=scaler.fit_transform(a)
```

```python
pd.DataFrame(a).head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(a,b,test_size=0.20,random_state=42)
```

```python
model=Sequential()
model.add(Dense(21,activation="relu"))
model.add(Dense(42,activation="relu"))
model.add(Dense(84,activation="relu"))
model.add(Dense(11,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy",optimizer="adam",metrics="accuracy")
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),batch_size=10,epochs=10,verbose=1)
model.summary()
```

```python
tahmin=model.predict(x_test)
```

```python
tahmin
```

```python
accuracy_score(y_test,model.predict(x_test))
```

```python
confusion_matrix(model.predict(x_test),y_test)
```
